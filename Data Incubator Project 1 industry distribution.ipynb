```python
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import matplotlib.axes as ax
from datetime import datetime 
from sympy import *
from matplotlib.dates import DateFormatter
from sklearn.datasets import fetch_mldata
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from collections import Counter
```

```python
# data visualization
data = pd.read_csv("temp_datalab_records_linkedin_company.csv", low_memory = False)
data = data[['company_name',"followers_count","employees_on_platform",'industry',"date_updated","date_added"]]
data = data.dropna(axis = 0)
data.describe()
print('Data dimension:',data.shape)
print('Number if industries:',len(np.unique(data.industry.astype(str))))
print('Industies:',np.unique(data.industry.astype(str)))
```

```python
Ukeys=np.unique(data.industry.astype(str))
#print(Ukeys)
#print((np.array(y_kmeans.shape)))
my_dict={}
for key in Ukeys:
   # print(key)
    temp1=data[data.industry == key]
    temp2=(temp1.company_name.values)[0]
   # print(len(temp2))
    my_dict[key]=len(temp2)

counts=Counter(my_dict)
n=len(my_dict)-1
topall = counts.most_common()[:-n-2:-1]
labelsall = []
sizesall = []
for i in range(len(topall)):
    labelsall.append(topall[i][0])
    sizesall.append(topall[i][1])
```

```python
#x = np.arange(len(list(counts.values())))
x = np.arange(len(topall[0:35]))
    #plt.bar(x, (counts.values()))
plt.figure(figsize=(2.5,12))
plt.barh(x,sizesall[0:35])
    #plt.xticks(x+.5, (counts.keys()))
plt.yticks(x+.5,labelsall[0:35])
plt.yticks(np.arange(len(x), step=1),rotation='horizontal')
plt.xlabel('Company Number')
plt.rcParams.update({'font.size': 16})
plt.axis([0,40,-1,35])
plt.show()

plt.figure(figsize=(2.5,12))
plt.barh(x,sizesall[35:70])
    #plt.xticks(x+.5, (counts.keys()))
plt.yticks(x+.5,labelsall[35:70])
plt.yticks(np.arange(len(x), step=1),rotation='horizontal')
plt.xlabel('Company Number')
plt.rcParams.update({'font.size': 16})
plt.axis([0,40,-1,35])
plt.show()
```
